Objective 1: Write down the PDF of $\mathcal N(1, 4)$.

This PDF is
$$
\frac{1}{2\sqrt{2\pi}}e^{-\frac{(x-1)^{2}}{8}}
$$

The graph of this PDF is as follows:

```
import numpy as np
import matplotlib.pyplot as plt #Import necessary packages
```

```
x = np.linspace(-6,8,280)
plt.plot(x, (1/(2*np.sqrt(2*np.pi))) * np.exp(-((x-1)**2)/8))
```
